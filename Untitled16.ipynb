```python
users = pd.read_csv("users.csv")
content = pd.read_csv("content.csv")
activity = pd.read_csv("user_activity.csv")
campaign = pd.read_csv("campaign.csv")
performance = pd.read_csv("campaign_performance.csv")
```

```python
print(users.head())
print(activity.info())
print(performance.isnull().sum())
```

```python
activity["Date"] = pd.to_datetime(activity["Date"])
```

```python
activity["Total_Engagement"] = activity["Watch_Time"] * activity["Sessions"]
```

```python
user_activity = activity.merge(users, on="User_ID")
```

```python
full_data = user_activity.merge(content, on="Content_ID")
```

```python
campaign_data = performance.merge(campaign, on="Campaign_ID")
```

```python
full_data.groupby("Region")["Total_Engagement"].sum().sort_values(ascending=False)
```

```python
full_data.groupby("Genre")["Watch_Time"].sum().sort_values(ascending=False)
```

```python
full_data.groupby("Subscription_Type")["Total_Engagement"].mean()
```

```python


# Merge user + activity
df = activity.merge(users, on="User_ID")

# Merge content
df = df.merge(content, on="Content_ID")

# Merge campaign performance
df = df.merge(performance, on="User_ID", how="left")

# Merge campaign details
df = df.merge(campaign, on="Campaign_ID", how="left")

# Create engagement column
df["Engagement"] = df["Watch_Time"] * df["Sessions"]

# Save final dataset
df.to_csv("final_tableau_dataset.csv", index=False)

print(" Final dataset created!")
```

```python
df["Conversion_Flag"] = df["Converted"].fillna(0)
df["Clicked_Flag"] = df["Clicked"].fillna(0)
```
